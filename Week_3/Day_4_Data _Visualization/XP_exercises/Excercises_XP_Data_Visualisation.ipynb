```python
import seaborn as sns
df = pd.read_csv("Student_Mental_health.csv")
sns.histplot(df["What is your CGPA?"], bins=20, color="yellow")
plt.title("Histogram of CGPA")
plt.show()
```

```python
df["Anxiety_num"]= df["Do you have Anxiety?"].map({"Yes":1, "No":0})

anxiety_by_gender = df.groupby("Choose your gender")["Anxiety_num"].mean().reset_index()

sns.barplot(x="Choose your gender", y="Anxiety_num", data=df, palette="Reds")
plt.title("Proportion of Students with Anxiety by Gender")
plt.ylabel("Proportion with Anxiety")
plt.show()
```

```python
df["Panic_num"] = df["Do you have Panic attack?"].map({"Yes":1, "No": 0})
panic_by_age = df.groupby("Age")["Panic_num"].mean().reset_index()

sns.lmplot(x="Age", y="Panic_num", data=df, aspect=1.5, height=5)
plt.title("Exploring the Relationship Between Age and Panic Attacks")
plt.show()
```

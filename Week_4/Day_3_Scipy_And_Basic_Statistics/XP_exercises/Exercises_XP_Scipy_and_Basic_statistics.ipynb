```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm
from scipy.stats import binom
from scipy.stats import pearsonr, spearmanr
```

```python
data = [12, 15, 13, 12, 18, 20, 22, 21]

scipy_mean = stats.tmean(data)
scipy_median = np.median(data)
scipy_variance = np.var(data)
scipy_std = np.std(data)

print(scipy_mean)
print(scipy_median)
print(scipy_variance)
print(scipy_std)
```

```python
 mu, sigma = 50, 10
 normal_dist = stats.norm(mu, sigma)

 x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
 pdf = normal_dist.pdf(x)

plt.figure(figsize=(8, 4))
plt.plot(x, pdf, label=f"Normal(μ={mu}, σ={sigma})")
plt.title("Normal Distribution (mean=50, std=10)")
plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

```

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_test = stats.ttest_ind(data1, data2)

print(t_test)
```

```python
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

predicted_price = slope * 90 + intercept
print(f"Predicted price for a house of size 90m²: {predicted_price}")
```

The slope is equal to 1828.7671232876712.
That means for every extra 1 m², the house price increases by 1828.7671232876712 currency units.

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("F-value:", f_value)
print("P-value:", p_value )
```

Based on the P-value, fertilizers don't have significant difference effect on plant growth as the data could be lucky.

If p-value were greater than 0.05, that mean the data has to much of a signifacnt difference for it to be taken likely

```python
n = 10
p = 0.5

prob_5 = binom.pmf(5, n, p)
print("P(X=5 heads out of 10:", prob_5)

k = np.arange(0, n+1)
probs = binom.pmf(k, n, p)
print("All probablilities 0..10:", probs.round(4))
print("They sum to:", probs.sum())

print("Mean (n*p):", n*p)
print("Variance (n*p*(1-p)):", n*p*(1-p))
```

```python
data = pd.DataFrame({
    'age':    [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

pearson_r, pearson_p = pearsonr(data["age"], data["income"])
print(f"Pearson r: {pearson_r:4f}, p-value: {pearson_p:4f}")

spearman_r, spearman_p = spearmanr(data["age"], data["income"])
print(f"Spearman rho: {spearman_r:4f}, p-value: {spearman_p:4f}")
```

# **3. Statistical Analysis with NumPy and SciPy:**

*   Execute detailed statistical analysis on each feature, including:
      Calculation of central tendency measures (mean, median, mode).
      Analysis of variability (range, variance, standard deviation).
      Evaluation of distribution shapes through skewness and kurtosis.
*   Perform hypothesis testing for statistical significance between groups (e.g., different price ranges).
*   Investigate feature-target correlations using SciPy.
*   Apply advanced SciPy statistical functions for deeper insights.









```python
means = df.mean()
medians = df.median()
modes = df.mode().iloc(0) #mode can return multiple rows, so we take one of them

ranges = df.max() - df.min()
variances = df.var()
stds = df.std()

skewness = df.skew()
kurtosis = df.kurtosis()

# Combine into one summary table
summary = pd.DataFrame({
    "Mean": means,
    "Median": medians,
    "Mode": modes,
    "Range": ranges,
    "Variance": variances,
    "StdDev": stds,
    "Skewness": skewness,
    "Kurtosis": kurtosis
})

print(summary.round(3))
```

```python
# -------- 1) Select numeric columns & compute skew/kurt --------
num_df = df.select_dtypes(include=[np.number]).copy()

skewness  = num_df.skew(numeric_only=True)
kurtosis_ = num_df.kurt(numeric_only=True)   # (excess kurtosis: 0 ~ normal)

summary = pd.DataFrame({
    "Mean":     num_df.mean(),
    "Median":   num_df.median(),
    "StdDev":   num_df.std(),
    "Range":    num_df.max() - num_df.min(),
    "Skewness": skewness,
    "Kurtosis": kurtosis_
}).round(3)

# Sort by absolute skew to see the most asymmetric first
summary_sorted = summary.reindex(skewness.abs().sort_values(ascending=False).index)

print("=== Skewness & Kurtosis Summary (numeric columns) ===")
display(summary_sorted)

# -------- 2) Pick columns to visualize --------

# cols_to_plot = ["ram", "battery_power", "price_range"]

# Automatically pick the top 3 most skewed columns:
cols_to_plot = list(skewness.abs().sort_values(ascending=False).head(3).index)

print(f"\nPlotting distributions for: {cols_to_plot}")

# -------- 3) Plot histograms + KDE with skew/kurt in the title --------
sns.set(style="whitegrid")
for col in cols_to_plot:
    plt.figure(figsize=(7,4))
    sns.histplot(num_df[col], kde=True, bins=30)
    plt.title(f"{col}  |  Skew={summary.loc[col,'Skewness']:.2f}  "
              f"Kurtosis={summary.loc[col,'Kurtosis']:.2f}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

# **4. Data Visualization with Matplotlib:**

*   Produce histograms, scatter plots, and box plots for data distribution and relationship insights.
*   Employ heatmaps for correlation visualization.
*   Ensure clarity in plots with appropriate titles, labels, and axis information.

```python
num_cols = df.select_dtypes("number").columns.tolist()

for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col); plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

```python
feat_list = ["ram", "battery_power", "px_height", "px_width", "mobile_wt"]

for col in feat_list:
    plt.figure(figsize=(6,3))
    sns.boxplot(data=df, x="price_range", y=col)
    plt.title(f"{col} by price_range")
    plt.xlabel("price_range"); plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

```python
pairs = [("px_width","px_height"), ("ram","battery_power"), ("mobile_wt","battery_power")]

for x,y in pairs:
    plt.figure(figsize=(6,3))
    sns.scatterplot(data=df, x=x, y=y, alpha=0.4)
    plt.title(f"{y} vs {x}")
    plt.tight_layout()
    plt.show()
```

```python
subset = df[["ram","battery_power","px_height","px_width","price_range"]]
sns.pairplot(subset, corner=True, diag_kind="hist")
```

```python
import numpy as np
num_df = df.select_dtypes('number')
corr = num_df.corr()

plt.figure(figsize=(9,7))
sns.heatmap(corr, cmap="coolwarm", center=0, annot=False, vmin=-1, vmax=1)
plt.title("Correlation heatmap (numeric features)")
plt.show()
```

**Insight Synthesis & Conclusion**

Key determinants of price:

RAM shows the strongest upward trend with price ranges  higher RAM consistently maps to higher prices.

Pixel resolution (px_height & px_width) also increases with price, making display quality another major driver.

Battery power contributes moderately; higher-end phones tend to carry larger batteries.

Less impactful features:

Binary flags like Bluetooth, Dual SIM, WiFi are evenly distributed across price ranges, offering little discriminative power.

Mobile weight shows weak/no clear link with price.

Statistical findings:

Distribution tests confirmed skewness in several features (e.g., front camera fc is right-skewed).

Hypothesis testing shows significant differences in mean RAM and resolution between price categories, validating them as strong predictors.

Unexpected observations:

Some high battery devices appear in lower price ranges  suggesting battery alone is not a premium driver.

Front camera (fc) and screen width (sc_w) have heavy skew/outliers, meaning theyre not reliable standalone signals.



---


The analysis highlights RAM, pixel resolution, and battery power as the primary determinants of price range classification.
Future modeling should prioritize these variables while treating features like connectivity options (Bluetooth, WiFi) as secondary.

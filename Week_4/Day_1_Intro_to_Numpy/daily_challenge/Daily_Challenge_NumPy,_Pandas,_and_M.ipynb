```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

daily_temperatures = np.random.uniform(low=-5, high=35, size=(10,12))

cities = ["Rishon", "Rehovot", "Ramat Gan", "Jaffa", "Hifa", "Tel Aviv", "Beer Sheva", "Jerusalem", "Eliat", "Ranana"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df = pd.DataFrame(daily_temperatures, index=cities, columns=months)
df.head(10)
```

```python
annual_average = pd.DataFrame.mean(df, axis=1)
annual_average
```

```python
hottest_city = annual_average.idxmax()
coldest_city = annual_average.idxmin()

print("Annual average temperature per city (°C):")
print(annual_average)
```

```python
print("Hottest city:", hottest_city, f"({annual_average[hottest_city]:}°C)")
print("Coldest city:", coldest_city, f"({annual_average[coldest_city]:}°C)")
```

```python
annual_average.plot(figsize=(12,6), linewidth=1.5, alpha=0.5)
plt.title("Annual average temperature per city")
plt.xlabel("months")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
# Plot with months on the x-axis, cities as lines
df.T.plot(figsize=(12,6), marker='o', linewidth=1.5)

plt.title("Monthly Temperature Trends for Each City", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.grid(True, alpha=0.3)
plt.legend(title="City", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.ylim(-5, 35)
plt.xticks(range(len(months)), months, rotation=45)
plt.tight_layout()
plt.show()
```
